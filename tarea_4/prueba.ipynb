```python
import sympy as sp


```

```python
# Definir las variables
x1, x2 = sp.symbols('x1 x2')

```

```python
# Definir la matriz H_f
H_f = sp.Matrix([
    [8*x1**2 + 4*(x1**2 + x2**2 - 1), 8*x1*x2],
    [8*x1*x2, 16*x2**2 + 4*(x1**2 + x2**2 - 1) + 4*(x2**2 - 1)]
])
display(H_f)
latex_code=sp.latex(H_f)
print(latex_code)
```

```python
ec_2=sp.Eq(x1,sp.sqrt(1-x2**2))
display(ec_2)
latex_code=sp.latex(ec_2)
print(latex_code)
```

```python
ec_3=sp.Eq(x1,sp.sqrt(2-x1**2)/sp.sqrt(2))
display(ec_3)
latex_code=sp.latex(ec_3)
print(latex_code)
```

```python
H_f_1=H_f.subs({x1: ec_2.rhs, x2:ec_3.rhs})
display(H_f_1)
latex_code=sp.latex(H_f_1)
print(latex_code)
```

```python
det_H_f = H_f_1.det()
display(det_H_f)
latex_code=sp.latex(det_H_f)
print(latex_code)
```
